```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD
from keras.utils import to_categorical, normalize
```

# Carica dati da csv

```python
df = pd.read_csv('SpotifyFeatures.csv')
```

```python
df
```

```python
# Features selection
df = df[['danceability',
         'energy',         
         'loudness',
         'speechiness',
         'acousticness',
         'instrumentalness',
         'liveness',
         'valence',
         'tempo',
         'duration_ms',
         'genre']].copy()

# Conversione in codici numerici
#df['key'] = df.key.astype('category').cat.codes
#df['time_signature'] = df.time_signature.astype('category').cat.codes
#df['artist_name'] = df.artist_name.astype('category').cat.codes
```

```python
df
```

```python
# Seleziona generi per il training
tgenres = {'Rock': 0, 'Classical': 1, 'Blues': 2, 'Dance': 3}

df_sel = df.loc[df['genre'].isin(tgenres.keys())].copy()

for genre in tgenres.keys():
    print(f'{genre} songs:', len(df_sel.loc[df['genre'] == genre]))
```

```python
# Converte le etichette del genere in valori numerici e.g 'Blues' -> 2
df_sel['genre'] = df_sel['genre'].apply(lambda x: tgenres[x])
```

# Rete neurale con Keras

## Preparazione dataset

```python
matrix = df_sel.values
np.random.shuffle(matrix) # Non necessario in quanto fit già prevede questa opzione
np.set_printoptions(precision=3, suppress=True)
print(matrix)
# Dimensioni test set (20% dataset)
data_set_size = np.size(matrix, 0)
test_set_size = (data_set_size * 20) // 100
print(f"\nTraining set size: {data_set_size - test_set_size}")
print(f"\nTest set size: {test_set_size}")
# Creazione training set
training_set = matrix[:-test_set_size]
# Creazione test set
test_set = matrix[-test_set_size:]
```

## Creazione modello

```python
model = Sequential()

model.add(Dense(100, activation='relu', input_dim=10))
model.add(Dropout(0.2))
model.add(Dense(50, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(25, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(4, activation='softmax'))
# Configura il modello per l'addestramento
opt = SGD(lr=0.05)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
# Training:
# training_set[...,:-1] prende i valori esclusa la colonna delle labels
x_train = training_set[...,:-1]

# min max normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
#print(scaler.fit(x_train))
#print(scaler.data_max_)
#print(scaler.transform(x_train))
x_train = scaler.fit_transform(x_train)

#Gaussian normalization
#meanv = np.mean(x_train, axis=0)
#stdv = np.std(x_train, axis=0)
#x_train = (x_train - meanv) / stdv
#from scipy.stats import zscore
#x_train = zscore(x_train)

print('Training set normalizzato:\n')
print(x_train)
# Converte le etichette in vettori one hot e.g. 1 -> [1, 0, 0, 0],  2 -> [0, 1, 0, 0]....
y_train = to_categorical(training_set[...,-1])
print('\nLabels oneHot:\n')
print(y_train)
```

```python
# Avvia l'addestramento del modello
#model.fit(x_train, y_train, epochs=3500, verbose=1, batch_size=20)
model.fit(x_train, y_train, batch_size=42, epochs=1500, verbose=2, validation_split=0.2, shuffle=True)
```

```python
#Testing
#x_test = normalize(test_set[...,:-1], axis=1, order=1)
x_test = test_set[...,:-1]
maxv = x_test.max(axis=0)
minv = x_test.min(axis=0)
x_test = (x_test - minv) / ( maxv - minv)

y_test = to_categorical(test_set[...,-1])

score = model.evaluate(x_test, y_test, batch_size=1)
print(f'\nAccuratezza (previsioni corrette / errate): {round(score[1] * 100, 3)}%')
print(f'\nEntità errore commesso per previsione errata: {round(score[0] * 100, 3)}%')
```

```python
# Salva il modello
while True:
    save = input("Salvare il modello? y/n > ")
    if save == 'y':
        model_name = input("Inserisci il nome >>> ")
        model.save(f'{model_name}.h5')
        break
    elif save == 'n':
        break
```
