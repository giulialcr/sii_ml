```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD
from keras.utils import to_categorical, normalize
```

# Carica dati da csv

```python
dataframe = pd.read_csv('SpotifyFeatures.csv')
```

```python
dataframe
#len(dataframe['artist_name'].unique()) -> 13318 artisti
```

```python
# Features selection
features = dataframe[['genre',
                      'danceability',
                      'energy',
                      'key',
                      'loudness',
                      'speechiness',
                      'acousticness',
                      'instrumentalness',
                      'liveness',
                      'valence',
                      'tempo',
                      'time_signature',
                      'duration_ms',
                      'artist_name',
                      'popularity']].copy()

# Conversione in codici numerici
features['key'] = features.key.astype('category').cat.codes
features['genre'] = features.genre.astype('category').cat.codes
features['time_signature'] = features.time_signature.astype('category').cat.codes
features['artist_name'] = features.artist_name.astype('category').cat.codes
```

```python
features
```

```python
# Converte i valori della popolarità in valori numerici associati ai range 0-10, 11-20, 21-30,...,91-100 
features['popularity'] = features['popularity'].apply(lambda x: int(x / 20.1))
sorted(features['popularity'].unique())
```

```python
# Statistiche popolarità
#popranges = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100']

for i in range(5):
    print(f'{int(i*20.1)}-{int((i+1)*20.1)}:', len(features.loc[features['popularity'] == i]))
```

```python
# Crea dataframe per addestramento
dataset = pd.DataFrame()
for i in range(5):
    subset = features.loc[features['popularity'] == i]
    dataset = dataset.append(subset.sample(2000, replace=True))

#subset.sample(408, replace=True)
#dataset
for i in range(5):
    print(f'{int(i*20.1)}-{int((i+1)*20.1)}:', len(dataset.loc[dataset['popularity'] == i]))
dataset
```

# Rete neurale con Keras

## Preparazione dataset

```python
matrix = features.values
np.random.shuffle(matrix) 
np.set_printoptions(precision=3, suppress=True)
print(matrix)
# Dimensioni test set (20% dataset)
data_set_size = np.size(matrix, 0)
test_set_size = (data_set_size * 20) // 100
print(f"\nTraining set size: {data_set_size - test_set_size}")
print(f"\nTest set size: {test_set_size}")
# Creazione training set
training_set = matrix[:-test_set_size]
# Creazione test set
test_set = matrix[-test_set_size:]
```

## Creazione modello

```python
model = Sequential()

model.add(Dense(140, activation='relu', input_dim=14))
model.add(Dropout(0.5))
model.add(Dense(70, activation='relu'))
#model.add(Dropout(0.3))
model.add(Dense(20, activation='relu'))
#model.add(Dropout(0.3))
model.add(Dense(5, activation='softmax'))
# Configura il modello per l'addestramento
opt = SGD(lr=0.05)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
# Training:
# training_set[...,:-1] prende i valori escludendo la colonna delle labels
x_train = training_set[...,:-1]
#x_train = training_set
# min max normalization
#maxv = x_train.max(axis=0)
#minv = x_train.min(axis=0)
#x_train = (x_train - minv) / ( maxv - minv)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
#print(scaler.fit(x_train))
#print(scaler.data_max_)
#print(scaler.transform(x_train))
x_train = scaler.fit_transform(x_train)



#l1-norm
#x_train = normalize(x_train, axis=0, order=1)

#Gaussian normalization
#meanv = np.mean(x_train, axis=0)
#stdv = np.std(x_train, axis=0)
#x_train = (x_train - meanv) / stdv
#from scipy.stats import zscore
#x_train = zscore(x_train)


print('Training set normalizzato:\n')
print(x_train)
# Converte le etichette in vettori one hot e.g. 1 -> [1, 0, 0, 0],  2 -> [0, 1, 0, 0]....
y_train = to_categorical(training_set[...,-1])
print('\nLabels oneHot:\n')
print(y_train)
```

```python
# Avvia l'addestramento del modello
#model.fit(x_train, y_train, epochs=3500, verbose=1, batch_size=20)

model.fit(x_train, y_train, batch_size=1000, epochs=200, verbose=2, validation_split=0.2, shuffle=True)
```

```python
#Testing
#x_test = normalize(test_set[...,:-1], axis=1, order=1)
x_test = test_set[...,:-1]
maxv = x_test.max(axis=0)
minv = x_test.min(axis=0)
x_test = (x_test - minv) / ( maxv - minv)

y_test = to_categorical(test_set[...,-1])

score = model.evaluate(x_test, y_test, batch_size=1)
print(f'\nAccuratezza (previsioni corrette): {round(score[1] * 100, 3)}%')
print(f'\nEntità errore commesso per previsione errata: {round(score[0] * 100, 3)}%')
```

```python
# Salva il modello
while True:
    save = input("Salvare il modello? y/n > ")
    if save == 'y':
        model_name = input("Inserisci il nome >>> ")
        model.save(f'{model_name}.h5')
        break
    elif save == 'n':
        break
```
